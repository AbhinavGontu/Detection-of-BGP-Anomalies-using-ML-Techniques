```python
import pandas as pd
```

# Import and Load Dataset

```python
code_redi_df = pd.read_csv("/content/drive/MyDrive/Dataset_CS240/Code_Red_I.csv")
slammer_df = pd.read_csv("/content/drive/MyDrive/Dataset_CS240/Slammer.csv")
nimda_df = pd.read_csv("/content/drive/MyDrive/Dataset_CS240/Nimda.csv")
```

```python
nimda_df.shape
#code_redi_df.shape
slammer_df.shape
```

# Combine the dataset

```python
dataset1 = pd.concat([slammer_df,nimda_df,code_redi_df],  ignore_index=True)
```

```python
X_train_1 = dataset1.iloc[:,:-1]
y_train = dataset1.iloc[:,-1]
X_train = X_train_1.drop(dataset1.columns[[0,1,2,3]], axis=1)
y_train = [0 if ele>0 else 1 for ele in y_train]
Y = pd.Series(y_train)
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X = sc.fit_transform(X_train)
```

```python
X_train.shape
print('Datatype:', type(X))
print('Datatype:', type(X_train))
#X_df = pd.DataFrame(data=X)
X_df = pd.DataFrame(data=X, columns = X_train.columns)
print('Datatype:', type(X_df))
X_df.head(10)
```

```python
from sklearn.model_selection import KFold 
```

```python
k = 10
kf = KFold(n_splits=k, random_state=None)
```

```python
from sklearn import svm
from sklearn import metrics
linear_svm_clf = svm.SVC(kernel="linear", probability=True)
rbf_svm_clf = svm.SVC(kernel="rbf")
cubic_svm_clf = svm.SVC(kernel="poly", degree=3, probability=True)
quad_svm_clf = svm.SVC(kernel="poly", degree=2, probability=True)
```

```python
acc_score_1 = []
acc_score_2 = []
acc_score_3 = []
acc_score_4 = []
f1_score_1 = []
f1_score_2 = []
f1_score_3 = []
f1_score_4 = []
X_df = pd.DataFrame(data=X)
print('Datatype:', type(X_df))
```

## Do not touch

```python
for train_index , test_index in kf.split(X):
    X_train , X_test = X_df.iloc[train_index,:],X_df.iloc[test_index,:]
    X_train.head(10)
    X_test.head(10)
    y_train , y_test = Y[train_index] , Y[test_index]
    y_train.head(10)
    y_test.head(10)
    linear_svm_clf.fit(X_train,y_train)
    pred_values = linear_svm_clf.predict(X_test)
     
    acc = metrics.accuracy_score(pred_values , y_test)
    acc_score.append(acc)
     
avg_acc_score = sum(acc_score)/k
print('accuracy of each fold - {}'.format(acc_score))
print('Avg accuracy : {}'.format(avg_acc_score))
```

# Accuracy and F1 score Linear kernel

```python
for train_index , test_index in kf.split(X_df):
    X_train , X_test = X_df.iloc[train_index,:],X_df.iloc[test_index,:]
    y_train , y_test = Y[train_index] , Y[test_index]
    linear_svm_clf.fit(X_train,y_train)
    pred_values = linear_svm_clf.predict(X_test)
     
    acc = metrics.accuracy_score(pred_values , y_test)
    acc_score_1.append(acc)
    f1 = metrics.f1_score(pred_values , y_test)
    f1_score.append(f1)
     
avg_acc_score = sum(acc_score_1)/k
avg_f1_score = sum(f1_score)/k
print('accuracy of each fold - {}'.format(acc_score_1))
print('Avg accuracy : {}'.format(avg_acc_score))
print('f1 score of each fold - {}'.format(f1_score))
print('Avg f1 score: {}'.format(avg_f1_score))
```
